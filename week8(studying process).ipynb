```python

from os.path import split
from sklearn.linear_model import Perceptron
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

```

```python

iris = datasets.load_iris()
x = iris.data
y = iris.target
print(x[:5])#hien thi ra 5 dong dau tien
print(y[:5])
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3) # lay ti le 70 /30 
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train_std[:5], x_train[:5])
print(y[:5])
```

```python
model = Perceptron(max_iter=1000000, eta0=0.2,random_state=0)
 #maxinter : so lan hoc
 #eta0 : toc do hoc 
 #randomstate : gia tri ngau nhien cua he thong
model.fit(x_train,y_train)
y_pred = model.predict(x_test_std)
print(y_pred)
print(y_test)
print("Độ chính xác của mô hình: ", accuracy_score(y_test,y_pred))
```

# CRABS

```python
import numpy as np
from sklearn.linear_model import Perceptron
#Chua data, co thu vien chuan hoa lan, chu, so viet tay, cac vat dung trong nha
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#Scale data de cho deu nhau (Ex:0, 1000 000 --> 0,1)
from sklearn.preprocessing import StandardScaler

import pandas as pd
url = 'crabs.csv'
data = pd.read_csv(url)

dataframe=data.replace(['F','M','B','O'],[0,1,1,0])
print(dataframe)
```

```python
x =dataframe.drop('speies', axis = 1)
y =dataframe.speies
print(x[:])
print(y[:])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train_std[:])
print(y.T)
```

```python
model = Perceptron(max_iter = 200, eta0= 0.13 , random_state = 0) #so lan train, toc do hoc
model.fit(x_train_std, y_train)
y_pred = model.predict(x_test_std)
print(y_pred[:])
print(y_test[:].T)
print("Do chinh xac:", accuracy_score(y_test, y_pred))
```

```python
x1 = [[0,50,23.1,20.2,46.2,52.2,21.1]]
y1 = model.predict(x1)
print(y1)
model.score(x1,y1)
```

# LUNG_CANCER



```python
import numpy as np
from sklearn.linear_model import Perceptron
#Chua data, co thu vien chuan hoa lan, chu, so viet tay, cac vat dung trong nha
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
#Scale data de cho deu nhau (Ex:0, 1000 000 --> 0,1)
from sklearn.preprocessing import StandardScaler

import pandas as pd
url1 = 'lung_cancer_examples.csv'
data = pd.read_csv(url1)
dataframe=data.drop('Name', axis = 1) #xoa cot
dataframe1=dataframe.drop('Surname', axis = 1)
x=dataframe1.drop('Result', axis = 1)
y =data.Result
print(x)
print(y)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
sc = StandardScaler()
sc.fit(x_train)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
print(x_train_std[:])
print(y[:])
model = Perceptron(max_iter = 200, eta0= 0.13 , random_state = 0) #so lan train, toc do hoc
model.fit(x_train_std, y_train)
y_pred = model.predict(x_test_std)
print(y_pred[:])
print(y_test[:].T)
print("Do chinh xac:", accuracy_score(y_test, y_pred))

```
