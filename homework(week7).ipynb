# bodyfat

```python

import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
url = 'bodyfat.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)

import seaborn as sns
sns.heatmap(correlation, annot = True)
```

```python

data.plot(kind = 'scatter', x = 'BodyFat', y = 'Density', figsize = (9,9), color = 'black')
```

```python

from sklearn import linear_model 
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['BodyFat']),y=data['Density'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['BodyFat']),y=data['Density'])
```

```python

data.plot(kind = 'scatter', x = 'BodyFat', y = 'Density', figsize = (9,9), color = 'black')
y_pred=model.predict(X=pd.DataFrame(data['BodyFat']))
plt.plot(data['BodyFat'],y_pred,color='red')
```

```python

from sklearn import linear_model 
model = linear_model.LinearRegression()
X=pd.DataFrame([data['BodyFat'],data['BodyFat']**2]).T
model.fit(X,y=data['Density'])
print('intercept cua mo hinh',model.intercept_)
print('confient cua mo hinh ',model.coef_)
print('do chinh xac cua mo hinh',model.score(X,y=data['Density']))
```

```python

import numpy as np
data.plot(kind = 'scatter', x = 'BodyFat', y = 'Density', figsize = (9,9), color = 'black', xlim = (0,50))
range = np.arange(0,6,0.1)
predictor = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```

# abalone

```python
import pandas as pd 
import numpy  as np
import matplotlib.pyplot as plt
url = 'abalone.csv'
data = pd.read_csv(url)
data.hist()
plt.show

correlation = data.corr()
print(correlation)

import seaborn as sns
```

```python

data.plot(kind = 'scatter', x = 'Length', y = 'Shell weight', figsize = (9,9), color = 'black')
```

```python

from sklearn import linear_model 
model=linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['Length']),y=data['Shell weight'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['Length']),y=data['Shell weight'])
```

```python
data.plot(kind = 'scatter', x = 'Length', y = 'Shell weight', figsize = (9,9), color = 'black')
y_pred=model.predict(X=pd.DataFrame(data['Length']))
plt.plot(data['Length'],y_pred,color='red')
```

```python
from sklearn import linear_model 
model = linear_model.LinearRegression()
X=pd.DataFrame([data['Length'],data['Length']**2,data['Length']**3,data['Length']**4]).T
model.fit(X,y=data['Shell weight'])
print('intercept cua mo hinh',model.intercept_)
print('confient cua mo hinh ',model.coef_)
print('do chinh xac cua mo hinh',model.score(X,y=data['Shell weight']))
```

```python

import numpy as np
data.plot(kind = 'scatter', x = 'Length', y = 'Shell weight', figsize = (9,9), color = 'black', xlim = (0,1))
range = np.arange(0,6,0.1)
predictor = pd.DataFrame([range, range**2,range**3, range**4]).T
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color = 'blue')
```
